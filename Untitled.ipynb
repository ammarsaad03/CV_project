```python
def load_dataset(images_path='images', annotations_path='annotations', target_size=(224, 224)):
    X = []
    y = []

    image_files = [f for f in os.listdir(images_path) if f.endswith('.png')]

    for img_file in image_files:
        xml_file = img_file.replace('.png', '.xml')
        xml_path = os.path.join(annotations_path, xml_file)

        if not os.path.exists(xml_path):
            continue

        # Load original image
        image_path = os.path.join(images_path, img_file)
        image = cv2.imread(image_path)
        orig_h, orig_w = image.shape[:2]

        # Resize and normalize
        preprocessed = preprocess_image(image, target_size)
        X.append(preprocessed)

        # Parse and normalize bounding boxes
        boxes = []
        tree = ET.parse(xml_path)
        root = tree.getroot()

        for obj in root.findall('object'):
            label = obj.find('name').text
            class_id = CLASSES.index(label)  # Convert label to integer

            bbox = obj.find('bndbox')
            x_min = int(bbox.find('xmin').text)
            y_min = int(bbox.find('ymin').text)
            x_max = int(bbox.find('xmax').text)
            y_max = int(bbox.find('ymax').text)

            # Normalize box to [0, 1] relative to original size
            x_min_norm = x_min / orig_w
            y_min_norm = y_min / orig_h
            x_max_norm = x_max / orig_w
            y_max_norm = y_max / orig_h

            boxes.append([class_id, x_min_norm, y_min_norm, x_max_norm, y_max_norm])

        y.append(boxes)

    return np.array(X), y
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
def build_object_detection_model(input_shape=(224, 224, 3), num_classes=3):
    inputs = layers.Input(shape=input_shape)

    # Feature extractor
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(inputs)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(128, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    # Output layer
    class_output = layers.Dense(num_classes, activation='softmax', name='class_output')(x)
    bbox_output = layers.Dense(4, activation='sigmoid', name='bbox_output')(x)  # [x_min, y_min, x_max, y_max]

    model = models.Model(inputs=inputs, outputs=[class_output, bbox_output])
    return model
```

```python
model = build_object_detection_model()
model.compile(
    optimizer='adam',
    loss={
        'class_output': 'sparse_categorical_crossentropy',
        'bbox_output': 'mse'  # bounding box regression
    },
    metrics={
        'class_output': 'accuracy',
        'bbox_output': 'mse'
    }
)
model.summary()
```

```python
def prepare_labels(y):
    """
    Convert y from list of lists to:
    - class_labels: shape (num_samples,)
    - bbox_labels: shape (num_samples, 4)
    
    Only uses the first object per image.
    """
    class_labels = []
    bbox_labels = []

    for boxes in y:
        if len(boxes) == 0:
            continue  # Skip images with no objects
        
        first_obj = boxes[0]
        class_id, x_min, y_min, x_max, y_max = first_obj
        class_labels.append(class_id)
        bbox_labels.append([x_min, y_min, x_max, y_max])
    
    return np.array(class_labels), np.array(bbox_labels)
```

```python
X, y = load_dataset()
class_labels, bbox_labels = prepare_labels(y)

print("Images:", X.shape)
print("Class labels:", class_labels.shape)
print("BBox labels:", bbox_labels.shape)
```

```python
model.fit(
    X,
    {
        'class_output': class_labels,
        'bbox_output': bbox_labels
    },
    epochs=20,
    batch_size=16,
    validation_split=0.2
)
```

```python
def visualize_predictions(model, X, class_labels, bbox_labels, num_samples=5):
    indices = random.sample(range(len(X)), num_samples)

    for idx in indices:
        image = X[idx]
        true_class = class_labels[idx]
        true_bbox = bbox_labels[idx]

        # Predict
        image_input = np.expand_dims(image, axis=0)
        pred_class_probs, pred_bbox = model.predict(image_input, verbose=0)
        pred_class = np.argmax(pred_class_probs[0])
        pred_bbox = pred_bbox[0]

        # Rescale boxes to image size
        img_disp = (image * 255).astype(np.uint8)
        h, w = img_disp.shape[:2]

        true_box = [int(true_bbox[0]*w), int(true_bbox[1]*h), int(true_bbox[2]*w), int(true_bbox[3]*h)]
        pred_box = [int(pred_bbox[0]*w), int(pred_bbox[1]*h), int(pred_bbox[2]*w), int(pred_bbox[3]*h)]

        # Draw boxes
        img_copy = img_disp.copy()
        cv2.rectangle(img_copy, (pred_box[0], pred_box[1]), (pred_box[2], pred_box[3]), (0,255,0), 2)
        cv2.putText(img_copy, f'Pred: {CLASSES[pred_class]}', (pred_box[0], pred_box[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

        cv2.rectangle(img_copy, (true_box[0], true_box[1]), (true_box[2], true_box[3]), (0,0,255), 2)
        cv2.putText(img_copy, f'True: {CLASSES[true_class]}', (true_box[0], true_box[1]+15), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)

        # Show image
        plt.figure(figsize=(6,6))
        plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title('Green = Prediction | Red = Ground Truth')
        plt.show()
```

```python
visualize_predictions(model, X, class_labels, bbox_labels, num_samples=3)
```
